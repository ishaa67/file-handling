```python
##43. Count above 75, below 35, between 35-75.

marks=np.array( [59, 61 ,35, 70, 82, 49, 53, 60, 69, 22, 73, 95, 98, 48, 99, 56, 68, 58, 68, 83, 85, 35, 63, 77
, 59, 97, 79, 87, 95, 90, 42, 59, 28, 80 ,29, 50, 23, 38, 85, 52, 72, 21, 22, 25, 21, 95, 93, 40,
 39, 91])

count_above_75 = np.sum(marks > 75)
below_35 = np.sum(marks < 35)
between_35to75 = np.sum((marks >= 35) & (marks <= 75))
#print
print("above 75:",count_above_75)
print("below 35:",below_35)
print("between 35-75:",between_35to75)
```

```python
#44. Assign grades.
marks = np.array([59, 61 ,35, 70, 82, 49, 53, 60, 69, 22, 73, 95, 98, 48, 99, 56, 68, 58, 68, 83, 85, 35, 63, 77
, 59, 97, 79, 87, 95, 90, 42, 59, 28, 80 ,29, 50, 23, 38, 85, 52, 72, 21, 22, 25, 21, 95, 93, 40,
 39, 91])
grades = ([])
for i in marks:
    if i >= 90:
        grades.append("A")
    elif i >= 75:
        grades.append("B")
    elif i >= 60:
        grades.append("C")
    elif i >= 40:
        grades.append("D")
    else:
        grades.append("F")
        
print(grades)
```

```python
arr = np.array([[89,67,87,90,67],
              [92,90,83,74,79],
              [82,84,92,91,89],
              [67,78,89,90,67],
              [90,78,78,90,67]])
#to find total
total = np.sum(arr,axis=1)
#to find average
average = np.average(arr,axis=1)
#to find percent
percentage = np.sum(arr,axis=1)/5

for i in range(len(total)):
    if i>= 90:
        grade = "A"
    elif i >= 75:
        grade = "B"
    elif i >= 60:
        grade = "C"
    elif i>= 40:
        grade = "D"
    else:
        grade = "F"

    print("Student", i + 1)
    print("Total      :", total[i])
    print("Average    :", average[i])
    print("Percentage :", percentage[i])
    print("Grade      :", grade)
    print()
```
